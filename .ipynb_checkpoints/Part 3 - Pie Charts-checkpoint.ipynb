```python
# Colors:
Blue = '#008fd5'
Red = '#fc4f30'
Yellow = '#e5ae37'
Green = '#6d904f'
```

```python
# plotting csv data
import pandas as pd
from collections import Counter
```

```python
data = pd.read_csv('data.csv')
ids, lang_response  = data['Responder_id'], data['LanguagesWorkedWith']

colors = [Blue, Red, Yellow, Green]

lang_counter = Counter()

for response in lang_response:
    lang_counter.update(response.split(';'))
    
languages, popularity = [], []

for item in lang_counter.most_common(5):
    languages.append(item[0]), popularity.append(item[1])
```

```python
explodes = [0, 0, 0, 0.1, 0]
plt.pie(popularity, labels=languages, explode=explodes,
        autopct='%1.f%%',shadow=True, 
        startangle=90, wedgeprops={'edgecolor':'black'})
plt.title('Top 5 Most Famous Programming Languages')
plt.tight_layout()
plt.show()
```
