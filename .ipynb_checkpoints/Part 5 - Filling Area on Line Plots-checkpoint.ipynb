```python
import pandas as pd
from matplotlib import pyplot as plt
# plt.style.use('ggplot')
```

```python
data = pd.read_csv('data_new.csv')
data.head()
```

```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Salary (USD)')
plt.plot(ages, dev_salaries, linestyle='--', color='k', label='All Devs')
plt.plot(ages, py_salaries, color='#444444', label='Python')
#plt.plot(ages, js_salaries, color='#e5ae37', label='JavaScript')
plt.tight_layout()

overall_median = 57287

plt.fill_between(ages, py_salaries, y2=dev_salaries,
                 where=(py_salaries > dev_salaries),
                 interpolate=True, alpha=0.25, label='Above Avg')

plt.fill_between(ages, py_salaries, y2=dev_salaries,
                 where=(py_salaries <= dev_salaries),
                 interpolate=True, color='red', alpha=0.25, label='Below Avg')

plt.legend()
#plt.grid(True)
plt.show()
```
