```python
data = pd.read_csv('2019-05-31-data.csv')
data.head()
```

```python
# plotting csv data
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, edgecolors='black',
            c=ratio, cmap='summer', linewidths=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Counts')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.show()
```
